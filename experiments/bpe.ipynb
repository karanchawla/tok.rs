```python
# Find the most common pair
def get_stats(tokens: list):
  counts = {}
  for i in range(len(tokens) - 1):
      k = (tokens[i], tokens[i + 1])
      counts[k] = counts.get(k, 0) + 1
  return counts

stats = get_stats(tokens)
print(stats)
print("Sorted stats:", sorted(((val, key) for key, val in stats.items()), reverse=True))
```

```python
top_pair = max(stats, key=stats.get)
top_pair
```

```python
# In the list of ids, find the pair and replace all occurences with replace
def merge(ids: list, pair: tuple, replace: int):
  new_ids = []
  i = 0
  while i < len(ids):
    if i < len(ids) -1 and ids[i] == pair[0] and ids[i + 1] == pair[1]:
      new_ids.append(replace)
      i += 2
    else:
      new_ids.append(ids[i])
      i += 1
  return new_ids

tokens2 = merge(tokens, top_pair, 256)
# print(tokens2)
# print("Length (token2): ", len(tokens2))
```

```python
# Desired vocab size
vocab_size = 276 # Hyper-parameter that we tune
num_merges = vocab_size - 256
ids = list(tokens)

merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  top_pair = max(stats, key=stats.get)
  ids = merge(ids, top_pair, 256 + i)
  merges[top_pair] = 256 + i

merges
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
  vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
  tokens = b"".join(vocab[idx] for idx in ids)
  text = tokens.decode('utf-8', errors='replace')
  return text
```

```python
def encode(string):
  # Given a string, return a list of integers
  tokens = string.encode('utf-8')
  while len(tokens) >=2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float('inf')))
    if pair not in merges:
      break # Nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)

  return tokens

print(encode("Hello World!"))
```
